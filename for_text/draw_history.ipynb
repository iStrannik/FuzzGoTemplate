```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Simple.csv', delimiter='\t')
x = df['epoch']
y = df['coverage']

plt.plot(x,y, marker='o')
plt.gcf().autofmt_xdate()

plt.show()
```

```python
df = pd.read_csv('depth.csv', delimiter=' ')
x = df['depth']
y = df['coverage']

plt.plot(x,y, marker='o')
plt.xlabel('depth')
plt.ylabel('coverage')
plt.gcf().autofmt_xdate()

plt.show()
```
