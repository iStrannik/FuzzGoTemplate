```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Simple.csv', delimiter='\t')
x = df['epoch']
y = df['coverage']

plt.plot(x,y, marker='o')
plt.gcf().autofmt_xdate()

plt.show()
```
